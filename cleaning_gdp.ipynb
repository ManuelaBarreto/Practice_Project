```python
import pandas as pd
```

```python
real_gdp =pd.read_csv('global/dataglobal/realgdpoecd.csv')
```

```python
real_gdp
```

```python
real_gdp["Multiplicateur d'unité"].unique()
```

```python
real_gdp.columns
```

```python
#dropping unimportant columns
real_gdp.drop(columns=[ "FREQ", "REF_AREA", "OBS_STATUS"], inplace=True)
```

```python
for column in real_gdp.columns:
    if (len(real_gdp[column].unique()) <= 1) and (column != "Unité de mesure"):
        real_gdp.drop(columns=[column],inplace=True)
```

```python
real_gdp
```

```python
real_gdp.drop(columns=["Fréquence d'observation", "TRANSACTION", "SECTOR", "REF_YEAR_PRICE",
                        "ACTIVITY", "EXPENDITURE", "TRANSFORMATION"], inplace=True)
```

```python
real_gdp
```

```python
real_gdp["Dépense"]
real_gdp.drop(columns=["Dépense"], inplace=True)
```

```python
real_gdp["Activité économique"].unique()
```

```python
real_gdp["Statut d'observation"].unique()
real_gdp.drop(columns=["Statut d'observation"], inplace=True)
```

```python
real_gdp
```

```python
real_gdp.to_csv("global/dataglobal/clean_realgdpg.csv")
```
