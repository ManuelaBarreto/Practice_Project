```python
import pandas as pd
import numpy as np
```

```python
def trimester_to_date(trimester):
    year, quarter = trimester.split('-')
    quarter=quarter[1]
    month = int(quarter) * 3
    return pd.Timestamp(year=int(year), month=month, day=1)
```

## Final Consumption Expenditure

```python
consumption = pd.read_csv("Global_original/finalconsumptionexpenditure.csv")
consumption.head()
```

```python
consumption.columns
```

```python
for column in consumption.columns:
    if (len(consumption[column].unique()) <= 1) and (column not in ["Multiplicateur d'unité", "Secteur institutionnel"]):
        consumption.drop(columns=[column],inplace=True)

consumption.head()
```

```python
consumption = consumption[consumption["FREQ"] == "Q"]
consumption = consumption[consumption["Ajustement"] == "Corrigé des variations saisonnières et des effets de calendrier"]
consumption = consumption[consumption["Type de prix"]=="Volumes chaînés"]
consumption = consumption[consumption["Transaction"]=="Dépense de consommation finale des résidents et non-résidents sur le territoire économique"]
```

```python
# Checking if the values of a country have the same reference
for i in list(consumption["Zone de référence"].unique()):
    if len(consumption[consumption["Zone de référence"] == i]['REF_YEAR_PRICE'].unique()) > 1:
        print(i)
```

```python
consumption.drop(columns=["Fréquence d'observation","FREQ","ADJUSTMENT","Ajustement","REF_AREA","TRANSACTION","Transaction","Type de prix","PRICE_BASE","OBS_STATUS","REF_YEAR_PRICE","Statut d'observation","Monnaie"], inplace= True)
```

```python
for column in consumption.columns:
    print(column)
```

```python
#Just checking that I still have all of the years

lista = sorted(list(consumption["TIME_PERIOD"].unique()))
lista_anos = []
for i in lista:
    try: 
        int(i)
        lista_anos.append(i)

    except:
        pass

for i in range(len(lista_anos)-1):
    if int(lista_anos[i+1]) != int(lista_anos[i]) + 1:
        print(lista_anos[i],lista_anos[i+1])
```

```python
for country in list(consumption['Zone de référence'].unique()):
    consumption.loc[consumption['Zone de référence']==country,"OBS_VALUE"] = consumption[consumption['Zone de référence']==country]["OBS_VALUE"].pct_change()
```

```python
consumption.TIME_PERIOD = consumption.TIME_PERIOD.apply(lambda x: trimester_to_date(x))
```

```python
consumption.head()
```

```python
consumption.to_csv("global/dataglobal/clean_finalconsumptionexpenditure.csv")
```

## GFCF

```python
for column in gfcf.columns:
    if (len(gfcf[column].unique()) <= 1) and (column not in ["Multiplicateur d'unité", "Secteur institutionnel"]):
        gfcf.drop(columns=[column],inplace=True)

gfcf.head()
```

```python
gfcf = gfcf[gfcf["FREQ"] == "Q"]
gfcf = gfcf[gfcf["Ajustement"] == "Corrigé des variations saisonnières et des effets de calendrier"]
gfcf = gfcf[gfcf["Type de prix"]=="Volumes chaînés"]
gfcf = gfcf[gfcf["Instruments financiers et actifs non financiers"].isin(['Actifs fixes, brut','Logements, brut'])]
```

```python
# Checking if the values of a country have the same reference
for i in list(gfcf["Zone de référence"].unique()):
    if len(gfcf[gfcf["Zone de référence"] == i]['REF_YEAR_PRICE'].unique()) > 1:
        print(i)
```

```python
gfcf.drop(columns=["Fréquence d'observation","FREQ","ADJUSTMENT","Ajustement","REF_AREA","INSTR_ASSET","Type de prix","PRICE_BASE","OBS_STATUS","REF_YEAR_PRICE","Statut d'observation","Monnaie"], inplace= True)
```

```python
for column in gfcf.columns:
    print(column)
```

```python
#Just checking that I still have all of the years

lista = sorted(list(gfcf["TIME_PERIOD"].unique()))
lista_anos = []
for i in lista:
    try: 
        int(i)
        lista_anos.append(i)

    except:
        pass

for i in range(len(lista_anos)-1):
    if int(lista_anos[i+1]) != int(lista_anos[i]) + 1:
        print(lista_anos[i],lista_anos[i+1])
```

```python
for country in list(gfcf['Zone de référence'].unique()):
    gfcf.loc[gfcf['Zone de référence']==country,"OBS_VALUE"] = gfcf[gfcf['Zone de référence']==country]["OBS_VALUE"].pct_change(periods=4)
```

```python
gfcf.TIME_PERIOD = gfcf.TIME_PERIOD.apply(lambda x: trimester_to_date(x))
```
